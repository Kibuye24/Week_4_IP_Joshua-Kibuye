## 1.5.5 Uniformity

```
# Procedure 1: None
# Data Cleaning:None
# Explanation:None
```

```
# Hint:To compute usage, we will need to understand we have to join successive (in time) measures/counters for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened.
# Creating a new column where we find the difference in the number of bluecars at the station/counter
#To make it uniform, we sort the table by days and by station.
df = df.sort_values(by=['day', 'Public name'])
#We then go ahead to create the new column
df['Bluecar_Diff'] = df['Bluecar counter'].diff()
df.head()
# Positive means a car was returned
# Negative or zero means a car was taken
```

# 1.6 Answering Research Questions

## 1.6.1 Bluecar counter 

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

df[['Bluecar counter']][df['City'] == 'Paris'][df['Bluecar_Diff'] <= 0].groupby(df['hour']).sum().sort_values(by='Bluecar counter',ascending=False)[0:3]
```

```
#What is the most popular hour for returning cars?

df[['Bluecar counter']][df['Bluecar_Diff'] > 0].groupby(df['hour']).sum().sort_values(by='Bluecar counter',ascending=False)[0:3]
```

```
#What station is the most popular? Overall?
df['Public name'].groupby(df['Public name']).count().sort_values(ascending=False)[0:1]
```

```
#What station is the most popular? At the most popular picking hour?
df.loc[df['hour'] == 6]['Public name'].groupby(df['Public name']).count().sort_values(ascending=False)[0:1]

   
```

```
# What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code? Overall?

df[['Bluecar counter']][df['Bluecar_Diff'] < 0].groupby(df['Postal code']).count().sort_values(by='Bluecar counter',ascending=False)[0:1]

#Paris/Porte de Montrouge/8 belongs to the 75014 postal code. It does not belong to the most popular postal code
  
    
```

```
# What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code? At the most popular picking hour?

df.loc[df['hour'] == 6][df['Bluecar_Diff'] < 0][['Bluecar counter']].groupby(df['Postal code']).count().sort_values(by='Bluecar counter',ascending=False)[0:1]
#Paris/Porte de Montrouge/8 belongs to the 75014 postal code. It does not belong to the most popular postal code
```
